```python
#Definimos nuestro intervalo, es decir, en donde vamos a estar evaluando los valores que toma Y
x = np.arange(-6,6)

y_1 = 3*x+5
y_2 = -x+3
y_3 = 2*x+1
```

```python
#Podemos ver que es un sistema SIN solucion ya que ninguna de sus variables se cruza en alguna oportunidad...
 #En este caso es porque nuestro sistema est sobre determinado, es decir, hay ms ecuaciones que variables

plt.figure()
plt.plot(x, y_1)
plt.plot(x, y_2)
plt.plot(x, y_3)

plt.axvline(x=0, color="grey")
plt.axhline(y=0, color="grey")
```

```python
#SISTEMA CON SOLUCIN

x = np.arange(-6,6)

y_2 = -1*x+3
y_3 = 2*x+1
```

```python
#Graficamos...
plt.figure()
plt.plot(x, y_2)
plt.plot(x, y_3)

plt.axvline(x=0, color="grey")
plt.axhline(y=0, color="grey")
plt.show()
```

```python
#INFINITAS SOLUCIONES
 #Cuando nos quedamos slo con 1 de las variables
    
x = np.arange(-6,6)

y_3 = 2*x+1
```

```python
#Es solo una recta... por lo que cualquiera podra ser la solucin

plt.figure()

plt.plot(x, y_3)
plt.axvline(x=0, color="grey")
plt.axhline(y=0, color="grey")
```
