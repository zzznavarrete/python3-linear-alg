```python
import numpy as np
```

```python
#Le pedimos a numpy que si tenemos un numero muy cercano a 0, que me muestre el 0 
np.set_printoptions(suppress=True)
```

```python
A = np.array([ [3,1], [2,1] ])
print(A)
```

```python
B = np.array([[1],[1]])
print(B)
```

```python
#Necesitamos... 1) Encontrar la inversa de A
inversa_A = np.linalg.inv(A)

print(inversa_A)
```

```python
#Nos da que nuestros valores deben ser.. 0 y 1
x = inversa_A.dot(B)
print(x)
```

```python
#Corroboramos que sea asi, realizando la operacin para ver si nos devuelve el vector que seria nuestro resultado (B)
print(A.dot(x))
```

```python
solucion_2 = inversa_A.dot(np.array([[3],[7]]))

print(solucion_2)
```

```python
#Corroborando la solucion
print(A.dot(solucion_2))
```
