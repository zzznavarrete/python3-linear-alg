```python
%matplotlib notebook

import numpy as np
import matplotlib.pyplot as plt
```

```python
v1 = np.array([1,1])
v2 = np.array([-1,-1])
```

```python
for a in range(-10,10):
    for b in range(-10,10):
        plt.scatter(v1[0]*a + v2[0]*b, v1[1]*a + v2[1]*b,
                   marker = ".",
                   color = "orange")
        
    
plt.xlim(-30,30)
plt.ylim(-30,30)

plt.axvline(x=0, color="grey")
plt.axhline(y=0, color="grey")

plt.show()
```

```python
#Lo que nos ocurrio es que vimos 2 vectores que son [1,1], [-1, -1]
 #Entonces, cuando hicimos las combinaciones lineales, lo Ãºnico que obtuvimos fue una recta
    #Esto en si mismo es un espacio, si nos abstraemos que estamos en r2 y solamente hubieramos entregado para graficar a X
    #Hubieramos obtenido lo mismo, en lugar de eso, lo que le entregamos fue elementos a r2 y obtuvimos un espacio (r1)
    #Es sÃ³lo una parte de R2
```

```python
v1 = np.array([1,0])
v2 = np.array([2,-3])
```

```python
for a in range(-10,10):
    for b in range(-10,10):
        plt.scatter(v1[0]*a + v2[0]*b, v1[1]*a + v2[1]*b,
                   marker = ".",
                   color = "orange")
        
    
plt.xlim(-30,30)
plt.ylim(-30,30)

plt.axvline(x=0, color="grey")
plt.axhline(y=0, color="grey")

plt.show()
```

```python
#Nos estÃ¡ dando R2.. es posible con estos vectores generar el espacio 
#Lo que vemos es que, con la combinaciÃ³n de ellos es posible escribir cualquier elemento de este espacio
#Cuando en el elemento anteriro NO podiamos, debido a que solo nos permitÃ­a movernos a travÃ©s de una recta

#En algebra lineal, siempre estamos trabajando en espacios vectoriales, lo cual no es mÃ¡s que nÃºmeros
#R, R1, R2, R3 -> Espacio escrito como vector
```

```python
#Ahora veamos como R2 en verdad aparte de ser un espacio es un sub espacio de R3
```

```python
from mpl_toolkits.mplot3d import Axes3D
```

```python
v1 = np.array([1,0,0])
v2 = np.array([2,-3,0])

fig = plt.figure()
ax = fig.add_subplot(111,projection="3d")

for a in range(-10,10):
    for b in range(-10,10):
        ax.scatter(v1[0]*a + v2[0]*b, v1[1]*a + v2[1]*b,
                   v1[2]*a + v2[2]*b,
                   marker = ".",
                   color = "orange")
        
ax.set_xlabel('Eje X')
ax.set_ylabel('Eje Y')
ax.set_zlabel('Eje Z')


plt.show()
```
