```python
import numpy as np
import matplotlib.pyplot as plt 
```

```python
x = np.arange(-5,5)
```

```python
y_1 = 3*x +5
y_2 = 2*x+3
```

```python
plt.figure()

plt.plot(x,y_1)
plt.plot(x, y_2)

plt.xlim(-5,5)
plt.ylim(-5,5)

plt.axvline(x=0, color="grey")
plt.axhline(y=0, color="grey")
```
